```python
#Compute rotations
def ocr(seq):
    seq = seq + '$'
    ocr_list = []
    for i in range(len(seq)):
        #print(seq[-i::])
        a = seq[:-i]
        b = seq[-i:]
        ocr_list.append(b + a)
        
    bwt = ''.join([x[-1] for x in sorted(ocr_list)])
    
    return(bwt)



```
