```python
#Compute rotations
def bwt(seq):
    seq = seq + '$' 
    rotations = [seq[i:] + seq[:i] for i in range(len(seq))]  # This line performs all the rotations
    sorted_rotations = sorted(rotations) # Sort the rotations
    bwt_transformation = ''.join(row[-1] for row in sorted_rotations)
    return(bwt_transformation)



```
